Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute:
Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:
Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
import numpy as np
import pandas as pd
from sklearn import datasets

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data  # Feature matrix as NumPy array
feature_names = iris.feature_names

# Calculate mean, median, standard deviation, and variance
means = np.mean(X, axis=0)
medians = np.median(X, axis=0)
std_devs = np.std(X, axis=0)
variances = np.var(X, axis=0)

# Identify which feature has the highest variability
variability = variances
highest_variability = feature_names[np.argmax(variability)]

# Convert one feature into shape (n,1)
sepal_length_reshaped = X[:, 0].reshape(-1, 1)

# Results
means, medians, std_devs, variances, highest_variability, sepal_length_reshaped.shape
```

Q6. Handling Missing Data (Critical ML Step)

Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain:
Why different strategies are used for different columns

```python

import pandas as pd
import seaborn as sns

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Fill missing age with median
titanic['age'] = titanic['age'].fillna(titanic['age'].median())

# Fill missing embarked with mode
titanic['embarked'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])

# Drop the deck column
titanic = titanic.drop(columns=['deck'])

# Identify missing values after filling
missing_values_after = titanic.isnull().sum()

# Results
missing_values_after
```

Q8. Group-Based Insights (Customer Segmentation Logic)

Context: ML models often depend on aggregated patterns.

Dataset: Tips

Tasks:

1. Compute:
Average tip by day
Total bill by gender
2. Create new feature:
tip_percentage
3. Identify:
Which day has highest tipping behavior
4. Explain:
How this helps in segmentation models

```python
# Load Tips dataset
tips = sns.load_dataset('tips')

# Compute average tip by day
average_tip_by_day = tips.groupby('day', observed=True)['tip'].mean()

# Total bill by gender
total_bill_by_gender = tips.groupby('sex', observed=True)['total_bill'].sum()

# Create new feature: tip_percentage
tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100

# Identify which day has the highest tipping behavior
highest_tip_day = average_tip_by_day.idxmax()

# Results
average_tip_by_day, total_bill_by_gender, highest_tip_day
```

Q11. End-to-End ML Preprocessing Pipeline (Titanic)

Context: Prepare dataset for ML model training

Tasks:

1. Load dataset
2. Handle missing values
3. Encode categorical variables (sex, embarked)
4. Normalize numerical features
5. Convert to NumPy array

Output:

Final feature matrix ready for model training

```python
# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Fill missing age with median
titanic['age'] = titanic['age'].fillna(titanic['age'].median())

# Fill missing embarked with mode
titanic['embarked'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])

# Drop the deck column
titanic = titanic.drop(columns=['deck'])

# Encode categorical variables (sex and embarked)
titanic = pd.get_dummies(titanic, columns=['sex', 'embarked'], drop_first=True)

# Normalize numerical features (age and fare)
titanic[['age', 'fare']] = (titanic[['age', 'fare']] - titanic[['age', 'fare']].mean()) / titanic[['age', 'fare']].std()

# Convert to NumPy array
final_feature_matrix = titanic.drop(columns='survived').to_numpy()

# Results shape
final_feature_matrix.shape
```

Q12. Feature Selection using Statistical Understanding

Context: Not all features improve model performance.

Dataset: Iris

Tasks:

1. Compute correlation matrix
2. Identify highly correlated features
3. Drop redundant features
4. Explain:
Impact on overfitting and model performance

```python
import numpy as np
import pandas as pd
from sklearn import datasets

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data  # Feature matrix as NumPy array
feature_names = iris.feature_names

# Compute correlation matrix
correlation_matrix = np.corrcoef(X, rowvar=False)

# Identify highly correlated features
threshold = 0.9
features_to_drop = []
for i in range(len(correlation_matrix)):
    for j in range(i+1, len(correlation_matrix)):
        if abs(correlation_matrix[i, j]) > threshold:
            features_to_drop.append(feature_names[j])  # Drop one of the correlated features

# Drop redundant features
# Assuming we keep features with lower index (earlier) over later ones
reduced_features = np.delete(X, [feature_names.index(f) for f in features_to_drop], axis=1)

# Results
features_to_drop
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
# Create age_group feature
def age_group(age):
    if age < 18:
        return 'Child'
    elif age < 65:
        return 'Adult'
    else:
        return 'Senior'

titanic['age_group'] = titanic['age'].apply(age_group)

# Compute survival rate per group
survival_rate_by_age = titanic.groupby('age_group')['survived'].mean()

# Results
survival_rate_by_age
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
# Generate random error values
errors = np.random.normal(0, 1, 1000)

# Compute mean and standard deviation
mean_errors = np.mean(errors)
std_errors = np.std(errors)

# Identify outliers
outliers = errors[(errors < mean_errors - 2 * std_errors) | (errors > mean_errors + 2 * std_errors)]

# Results
mean_errors, std_errors, outliers.shape
```
