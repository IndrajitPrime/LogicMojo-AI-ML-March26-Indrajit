Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

Write a Python program that performs the following analysis:
1. Identify and display all transaction values that appear more than once.
2. Detect transactions greater than ₹5000 as these may require fraud monitoring.
3. Calculate the total spending for the day.
4. Create a new list containing only unique transaction values.
5. Sort the unique transaction list in descending order to view the largest transactions first.

```python
transactions = [1200,3000,250,7800,540,3000,9100,250,1500,7800]
total = 0
new_list = []
for i in range(len(transactions)):
#1
    if(transactions.count(transactions[i])>1):
        print("Value =", transactions[i])
        print("Count =", transactions.count(transactions[i]))
#2
    if(transactions[i]>5000):
        print("Fraud Check Required! Value=",transactions[i])
#3 Calculate the total. Print outside loop
    total = total + transactions[i]
#4
    if transactions[i] not in new_list:
        new_list.append(transactions[i])
    
#5
print("New List=", new_list)
new_list.sort(reverse=True)
print("Sorted Descending :",new_list)
print("Total Spending =",total)
```

Question 2:- A dataset exported from a database contains the following customer age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
Some entries are invalid because they are either negative numbers or missing values (None).

Write a Python program that:
1. Iterates through the dataset and removes all invalid age values.
2. Stores only valid ages in a new list.
3. Calculates the average age of valid customers.
4. Displays the list of customers whose age is greater than 30.
5. Prints how many invalid entries were removed during the cleaning process.

```python
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
valid_ages=[]
new_list=[]
above_thirty=[]
count=0
total=0
avg=0
for age in range(len(ages)):
    valid_ages.append(ages[age])
#1    
for i in range(len(valid_ages)-1,-1,-1):
    if valid_ages[i] is None or valid_ages[i]<0:
        valid_ages.pop(i)
#5
        count=count+1 
#2
for i in range(len(valid_ages)):
  if valid_ages[i] is not None and valid_ages[i]>0:
      new_list.append(valid_ages[i])
for j in range(len(new_list)):
    if new_list[j]>30:
#4
        above_thirty.append(new_list[j])
#3
    total=total+new_list[j]
avg=total/len(new_list)
print("List of invalid ages removed: valid_ages=",valid_ages)
print("New List with valid ages only: new_list=",new_list)
print("Average valid age =",avg)
print("List of ages above 30: above_thirty=",above_thirty)
print("Count of invalid ages removed =",count)
        

        
    
```

Question 3:- An analyst wants to compare the price of products listed on two e-commerce platforms.

The following data is provided:
products = ["Laptop","Phone","Tablet","Monitor","Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

Write a Python program that:
1. Combines the product names and their prices into a single data structure representing a table.
2. Compares the price of each product across both platforms.
3. Determines which platform offers the lower price.
4. Calculates the difference between the two prices.
5. Displays the product name, cheaper platform, and price difference.

```python
products = ["Laptop","Phone","Tablet","Monitor","Headphones"] 
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]
cheap=[]
diff=[]
combined=[]
for p,a,f in zip(products,amazon_prices,flipkart_prices):
    combined.append([p,a,f])
print(combined)
#1
print("products,amazon price,flipkart price")
for i in range(len(combined)):
    print(combined[i])
#2
for i in range(len(combined)):
    if(combined[i][1]==combined[i][2]):
        print(f"For Product {combined[i][0]} Prices are equal")
        cheap.append("Same")
        diff.append(0)
        
    elif(combined[i][1]>combined[i][2]):
        print(f"For Product {combined[i][0]} Amazon Price is greater")
        cheap.append("Flipkart")
        diff.append(combined[i][1]-combined[i][2])
    else:
        print(f"For Product {combined[i][0]} Flipkart Price is greater")
        cheap.append("Amazon")
        diff.append(combined[i][2]-combined[i][1])
for j in range(len(combined)):
    print(f"Product = {combined[j][0]}, Cheaper Platform = {cheap[j]}, Price Difference = {diff[j]}.")
   
```

Question 4:- A web platform records the hours when users log into the system.

login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
Write a Python program that:

1. Separates login hours into even-hour logins and odd-hour logins.
2. Identifies login hours that fall into peak activity time (hours greater than 5).
3. Sorts the login hours in ascending order.
4. Calculates how many logins occurred during peak hours.

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
even_login=[]
odd_login=[]
peak_login=[]
count=0
for i in range(len(login_hours)):
    if login_hours[i] > 5:
        count=count+1
        peak_login.append(login_hours[i])
    if login_hours[i]%2 == 0:
        even_login.append(login_hours[i])
    else:
        odd_login.append(login_hours[i])
login_hours.sort()
even_login.sort()
odd_login.sort()

print("IDentified peak logins: peak_login=",peak_login)
print("Sorted original login hours: login_hours=",login_hours)
print("Sorted even login hours: even_login=",even_login)
print("Sorted odd login hours: odd_login=",odd_login)
print("Total peak logins: count = ",count)

    
```
